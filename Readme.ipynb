### What is the difference between characterTextSpliiter and RecursiveCharacterTextSplitter in the langchain library? 

## CharacterTextSplitter
- Splits text using a single fixed separator (e.g., \n\n, \n, space).Simple and fast.May break sentences or paragraphs awkwardly.

## RecursiveCharacterTextSplitter
- Splits text recursively using multiple separators (e.g., paragraphs → lines → sentences → characters).Preserves semantic structure better.Preferred for LLM embeddings & RAG use cases.
