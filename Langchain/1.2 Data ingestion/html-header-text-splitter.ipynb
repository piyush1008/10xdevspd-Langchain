### How to split JSON data

This json splitter split json data while allwoiing cotol over hunk sizes. It traverses json data depth first and builds smaller json chunks. It attempts to keep nested json objects whole but will split the m if nneded to keep chunks between a min_chunk_size and the max_chunk_size.

If the valiue is not as nested json, buit rather a very large string will string will not be split. It you need a har cap on the chunk size consider composing this with a Recursizve text splitters on thos chunks. 

```python
import json
import requests

json_data=requests.get("https://api.smith.langchain.com/openapi.json").json()
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

json_splitter=RecursiveJsonSplitter(max_chunk_size=300)

json_chunks=json_splitter.split_json(json_data)
json_chunks
```

```python
for chunk in json_chunks[:3]:
    print(chunk)
```

```python
## the splitter can also create document

docs=json_splitter.create_documents(texts=[json_data])
for doc in docs[:3]:
    print(doc)
```
