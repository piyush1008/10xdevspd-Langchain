### How to split by character-character text splitter

This is the simplest method. This split based on a given character sequence, which default to "\n\n". Chunk lenght is measured by number of characters.

    1. How the text is split: by single character seperator
    2. How the chunk size is measured: by number of characters.

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader("speech.txt")

docs=loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n", chunk_size=10, chunk_overlap=3)
text_splitter.split_documents(docs)


```

```python
speech=""
with open("speech.txt","r") as file:
    speech=file.read()

text_splitter=CharacterTextSplitter(separator="\n", chunk_size=10, chunk_overlap=3)
text=text_splitter.create_documents([speech])
print(text[0])
print(text[1])
```
