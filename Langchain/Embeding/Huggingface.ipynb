### Sentence transformers on Hugging face

Python framework for state-of-the-art sentence, text and image embeddings. ONe of the mebedding model is used in the HuggingFaceEmbeddings class. We have also added an alias for  SentenceTransformersEmbeddings for users who are familar with directly using that package

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="this is a test document"
query_result=embedding.embed_query(text)
query_result


# doc_result=embedding.embed_documents([text, "This is not a test document"])
# doc_result
```
