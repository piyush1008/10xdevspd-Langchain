### Embedding Technique

Converting text into vectors
An embedding is a vector (list) of floating point numbers. The distance between two vectors measures their relatedness. Small distances suggest high relatedness and large distances suggest low relatedness.

OpenAI embedding models-
https://platform.openai.com/docs/guides/embeddings


```python
import os 
from dotenv import load_dotenv
load_dotenv()   ##load the env variable

os.environ["OPEN_API_KEY"]=os.getenv("OPEN_API_KEY")


from langchain_openai import OpenAIEmbeddings
embedding=OpenAIEmbeddings(model="text-embedding-3-large")
embedding

text="this is tutorial on the openai embedding"

query_result=embedding.embed_query(text)

query_result

```

```python
embedding_1024=OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
embedding_1024

text="this is tutorial on the openai embedding"

query_result=embedding.embed_query(text)

query_result
```

```python
## data ingestion
from langchain_community.document_loaders import TextLoader
from langchain_openai import OpenAIEmbeddings
loader=TextLoader("speech.txt")

docs=loader.load()
docs

# data Transformation , converting them into chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=8, chunk_overlap=4)
final_document=text_splitter.split_documents(docs)


final_document

# convert it into vector using openAI embedding  and store those in vectore store
embedding_1024=OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
embedding_1024
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_document, embedding_1024)

db

# querying the vectore store 
query="You should be"
retrieved_result=db.similarity_search(query)
print(retrieved_result)
```
