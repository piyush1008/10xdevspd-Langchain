### Ollama
Ollama support emebdding models , making it possiblel to build RAG application that combine text prompts with existing documents or other data

https://ollama.com/blog/embedding-models

```python
from langchain_community.embeddings import OllamaEmbeddings
embedding=OllamaEmbeddings(model="gemma2:2b")

r1=embedding.embed_documents(
    ["Alpha is the first character of the greek alphabet",
    ["Beta is the second character of the greek alphabet"]
    ]
)
print(r1)
len(r1[0])
```

```python
r2=embedding.embed_query("What is the second letter of Grek Alphabet")
r2


```
