### Chroma

It is a AI-native open-source vector database focused on developer productivity and happiness

https://docs.langchain.com/oss/python/integrations/vectorstores

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

document=TextLoader("speech.txt").load()


text_splitter=RecursiveCharacterTextSplitter(chunk_size=10, chunk_overlay=8)
docs=text_splitter(document)

embedding=OllamaEmbeddings(model="gemma2:2b")

vectordb=chroma.from_documents(documents=docs, embedding=embedding)



query="what is the main reson from trump attacked Iraq?"
doc=vectordb.similarity_search(query)
doc[0].page_content
```

```python
## Saving the disk 
vectordb=Chroma.from_documents(documents=docs, embedding=embedding, persist_directory="./chroma_db")
```

```python
## load from the disk
db2=Chroma(perist_directory="./chroma_db", embedding_function=embedding)
docs=db2.similarity_search(query)
print(docs[0].page_content)
```

```python
###Retriever 

retriever=vectordb.as_retriever()
retriever.invoke(query)[0].page_content
```
